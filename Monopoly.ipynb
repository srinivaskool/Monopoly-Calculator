```python
from tkinter import *
import sqlite3

def create():
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS calculator (id INTEGER PRIMARY KEY,player_1 TEXT,player_2 TEXT,player_3 TEXT,player_4 TEXT,cash_1 FLOAT,cash_2 FLOAT,cash_3 FLOAT,cash_4 FLOAT,mort_1 FLOAT,mort_2 FLOAT,mort_3 FLOAT,mort_4 FLOAT)")
    conn.commit()
    conn.close()

create()

def insert():
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("INSERT INTO calculator VALUES(NULL,'Player 1','Player 2','Player 3','Player 4',150,150,150,150,0,0,0,0)")
    conn.commit()
    conn.close()
    
#insert()

def delete(id):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("DELETE FROM calculator WHERE id = ?",(id,))
    conn.commit()
    conn.close()

#delete(3)

def view():
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    conn.commit()
    conn.close()
    return rows

#print(view())

def update_1(cash_1):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    real_cash_1 = rows[0][5] + cash_1
    cur.execute("UPDATE calculator SET cash_1= ? WHERE player_1= ?",(real_cash_1,"Player 1"))
    conn.commit()
    conn.close()
    return rows

def update_2(cash_2):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator WHERE player_2 = 'Player 2'")
    rows = cur.fetchall()
    real_cash_2 = rows[0][6] + cash_2
    cur.execute("UPDATE calculator SET cash_2= ? WHERE player_2= ?",(real_cash_2,"Player 2"))
    conn.commit()
    conn.close()
    return rows

def update_3(cash_3):
    conn = sqlite3.connect("monopoly.db"
                          )
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator WHERE player_3 = 'Player 3'")
    rows = cur.fetchall()
    real_cash_3 = rows[0][7] + cash_3
    cur.execute("UPDATE calculator SET cash_3= ? WHERE player_3= ?",(real_cash_3,"Player 3"))
    conn.commit()
    conn.close()
    return rows

def update_4(cash_4):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator WHERE player_4 = 'Player 4'")
    rows = cur.fetchall()
    real_cash_4 = rows[0][8] + cash_4
    cur.execute("UPDATE calculator SET cash_4= ? WHERE player_4= ?",(real_cash_4,"Player 4"))
    conn.commit()
    conn.close()
    return rows

def show_update_cash():
    tc1.delete(0,END)
    tc2.delete(0,END)
    tc3.delete(0,END)
    tc4.delete(0,END)
    update_1(c1_text.get())
    update_2(c2_text.get())
    update_3(c3_text.get())
    update_4(c4_text.get())
    r = view()
    tc1.insert(END,r[0][5])
    tc2.insert(END,r[0][6])
    tc3.insert(END,r[0][7])
    tc4.insert(END,r[0][8])
    c1.delete(0,END)
    c1.insert(END,"0")
    c2.delete(0,END)
    c2.insert(END,"0")
    c3.delete(0,END)
    c3.insert(END,"0")
    c4.delete(0,END)
    c4.insert(END,"0")

def mort_update_1(mort_1):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    real_mort_1 = rows[0][9] + mort_1
    cur.execute("UPDATE calculator SET mort_1= ? WHERE player_1= ?",(real_mort_1,"Player 1"))
    conn.commit()
    conn.close()
    return rows

def mort_update_2(mort_2):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    real_mort_2 = rows[0][10] + mort_2
    cur.execute("UPDATE calculator SET mort_2= ? WHERE player_2= ?",(real_mort_2,"Player 2"))
    conn.commit()
    conn.close()
    return rows

def mort_update_3(mort_3):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    real_mort_3 = rows[0][11] + mort_3
    cur.execute("UPDATE calculator SET mort_3= ? WHERE player_3= ?",(real_mort_3,"Player 3"))
    conn.commit()
    conn.close()
    return rows

def mort_update_4(mort_4):
    conn = sqlite3.connect("monopoly.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM calculator")
    rows = cur.fetchall()
    real_mort_4 = rows[0][12] + mort_4
    cur.execute("UPDATE calculator SET mort_4= ? WHERE player_4= ?",(real_mort_4,"Player 4"))
    conn.commit()
    conn.close()
    return rows

def show_update_mort():
    tm1.delete(0,END)
    tm2.delete(0,END)
    tm3.delete(0,END)
    tm4.delete(0,END)
    mort_update_1(m1_text.get())
    mort_update_2(m2_text.get())
    mort_update_3(m3_text.get())
    mort_update_4(m4_text.get())
    r = view()
    tm1.insert(END,r[0][9])
    tm2.insert(END,r[0][10])
    tm3.insert(END,r[0][11])
    tm4.insert(END,r[0][12])
    m1.delete(0,END)
    m1.insert(END,"0")
    m2.delete(0,END)
    m2.insert(END,"0")
    m3.delete(0,END)
    m3.insert(END,"0")
    m4.delete(0,END)
    m4.insert(END,"0")

def button_function():
    show_update_mort()
    show_update_cash()
    nn1.delete(0,END)
    nn1.insert(END,n1_text.get())
    nn2.delete(0,END)
    nn2.insert(END,n2_text.get())
    nn3.delete(0,END)
    nn3.insert(END,n3_text.get())
    nn4.delete(0,END)
    nn4.insert(END,n4_text.get())

window = Tk()
window.wm_title("Monopoly Calculator")

l1 = Label(window,text="Player Name")
l1.grid(row=0,column=0)

l2= Label(window,text="Cash Transaction")
l2.grid(row=0,column=1)

l3= Label(window,text="Mortage Transaction")
l3.grid(row=0,column=2)

n1_text = StringVar()
n1 = Entry(window,textvariable = n1_text)
n1.insert(END,"Player 1")
n1.grid(row=1,column=0)

n2_text = StringVar()
n2 = Entry(window,textvariable = n2_text)
n2.insert(END,"Player 2")
n2.grid(row=2,column=0)

n3_text = StringVar()
n3 = Entry(window,textvariable = n3_text)
n3.insert(END,"Player 3")
n3.grid(row=3,column=0)

n4_text = StringVar()
n4 = Entry(window,textvariable = n4_text)
n4.insert(END,"Player 4")
n4.grid(row=4,column=0)

c1_text = IntVar()
c1 = Entry(window,textvariable = c1_text)
c1.grid(row=1,column=1)

c2_text = IntVar()
c2 = Entry(window,textvariable = c2_text)
c2.grid(row=2,column=1)

c3_text = IntVar()
c3 = Entry(window,textvariable = c3_text)
c3.grid(row=3,column=1)

c4_text = IntVar()
c4 = Entry(window,textvariable = c4_text)
c4.grid(row=4,column=1)

m1_text = IntVar()
m1 = Entry(window,textvariable = m1_text)
m1.grid(row=1,column=2)

m2_text = IntVar()
m2 = Entry(window,textvariable = m2_text)
m2.grid(row=2,column=2)

m3_text = IntVar()
m3 = Entry(window,textvariable = m3_text)
m3.grid(row=3,column=2)

m4_text = IntVar()
m4 = Entry(window,textvariable = m4_text)
m4.grid(row=4,column=2)

b1 = Button(window,text="Submit",width=16,background="#FFA500",command=button_function)
b1.grid(row=5,column=1)

l1 = Label(window,text="Player Name")
l1.grid(row=6,column=0)

l2= Label(window,text="Total Cash")
l2.grid(row=6,column=1)

l3= Label(window,text="Total Mortage")
l3.grid(row=6,column=2)

nn1 = Entry(window)
nn1.grid(row=7,column=0)

nn2 = Entry(window)
nn2.grid(row=8,column=0)

nn3 = Entry(window)
nn3.grid(row=9,column=0)

nn4 = Entry(window)
nn4.grid(row=10,column=0)

tc1 = Entry(window)
tc1.grid(row=7,column=1)

tc2 = Entry(window)
tc2.grid(row=8,column=1)

tc3 = Entry(window)
tc3.grid(row=9,column=1)

tc4 = Entry(window)
tc4.grid(row=10,column=1)

tm1 = Entry(window)
tm1.grid(row=7,column=2)

tm2 = Entry(window)
tm2.grid(row=8,column=2)

tm3 = Entry(window)
tm3.grid(row=9,column=2)

tm4 = Entry(window)
tm4.grid(row=10,column=2)

window.mainloop()
```
